## Question 1:

### Create an empty list. Accept 10 numbers from the user and append to it the list if it is an even number.

```python
# Creating an empty list
l = []

# Accepting 10 numbers from the user
for i in range(10):
    x = int(input())
    if x%2==0:
        l.append(x)    # Appending only even numbers
        
print(*l,end=' ')
```

## Question 3:

### Given a number n, write a program that generates a dictionary d which contains (i, i*i), where i is from 1 to n (both included). Print this dictionary d.

```python
# Creating an empty dictionary
d = {}

#  Getting the value for n from the user
n = int(input())

for i in range(1,n+1):
    d[i] = i*i

print(d)
```

## Question 4:

### Write a program to compute the distance between the current position after a sequence of movement and original point. If the distance is a float, then just print the nearest integer (use round() function for that and then convert it into an integer).

```python
#  Creating an origin point
pos = {"x":0,"y":0}

# Getting the number of directions from the user
n = int(input())

# for loop
for i in range(n):
    move =  input().split(" ")      # Accepting Movements and Storing it as a list
    
    if move[0].lower() == "up":     # Extracting direction and comparing
        pos["y"] += int(move[1])    # Increment/Decrement Appropriate Co-Ordinates
    
    elif move[0].lower() == "down":
        pos["y"] -= int(move[1])
    
    elif move[0].lower() == "left":
        pos["x"] -= int(move[1])
    
    elif move[0].lower() == "right":
        pos["x"] += int(move[1])

#  Printing the result

print(int(round((pos["x"]**2 + pos["y"]**2)**0.5)))   # DISTANCE FROM ORIGIN
```
