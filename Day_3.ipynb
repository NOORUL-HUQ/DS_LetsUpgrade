## Question 2:

### Accept two lists of 5 elements each from the user. Convert them to numpy arrays. Concatenate these arrays and print it. Also sort these arrays and print it.

```python
# Importing Numpy with alias name
import numpy as np

# Creating 2 empty lists:
lst_1 = []
lst_2 = []

# Getting values for list_1 from user:
for i in range(5):
    x=int(input())
    lst_1.append(x)

# Getting values for list_2 from user:
for i in range(5):
    x=int(input())
    lst_2.append(x)

# Converting list into array:
arr_1 = np.array(lst_1)
arr_2 = np.array(lst_2)

# Concatenating 2 array:
arr = np.concatenate((arr_1, arr_2))
print(arr)

# sorting up concatenated array
sortArr = np.sort(arr)
print(sortArr)
```

## Question 4:

### How to convert a 1D array into a 2D array? Demonstrate with the help of a code snippet

```python
# importing numpy
import numpy as np
x = np.array([1, 2])

#ArrayName[numpy.newaxis, :]               - row major
#ArrayName[:, numpy.newaxis]               - column major
#numpy.expand_dims(ArrayName, axis = 0)    -  row major
#numpy.expand_dims(ArrayName, axis = 1)    -  column major

m = arr[np.newaxis, :]
n = arr[:, np.newaxis]
p = np.expand_dims(x, axis=0)
q = np.expand_dims(x, axis=1)
r = np.expand_dims(x, axis=(0, 1))
s = np.expand_dims(x, axis=(2, 0))
print('\n',m) # converting using newaxis in row wise
print('\n',n) # converting using newaxis in column wise
print('\n',p) # converting using newaxis in row wise
print('\n',q) # converting using newaxis in column wise
print('\n',r)
print('\n',s)
```

## Question 5:

### Consider two square numpy arrays. Stack them vertically and horizontally.

```python
# importing numpy
import numpy as np

# creating 2 arrays with square values
arr_1 = np.square([1, 2, 3, 4, 5])
arr_2 = np.square([6, 7, 8, 9, 10])

# stacking and printing the value
print(np.hstack((arr_1, arr_2))) # Horitzontal Stack
print(np.vstack((arr_1, arr_2))) # Vertical Stack
```
